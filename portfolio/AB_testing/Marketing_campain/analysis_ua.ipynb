# A/B Тестування: Вплив реклами на конверсію

Цей аналіз оцінює, чи має рекламна кампанія статистично значущий вплив на конверсію користувачів. Мета — визначити, чи користувачі, які бачили рекламу (експериментальна група), конвертуються частіше, ніж ті, хто бачив лише соціальну рекламу (PSA, контрольна група).

## **Гіпотези**

- **Нульова гіпотеза (H0)**:  
  Конверсія у групі, що бачила рекламу (`ad`), дорівнює конверсії у контрольній групі (`psa`).
  
- **Альтернативна гіпотеза (H1)**:  
  Конверсія у групі `ad` є вищою, ніж у групі `psa`.

## **Огляд даних**
Датасет містить інформацію про взаємодію користувачів із рекламою, зокрема:
- Унікальний ідентифікатор користувача
- Приналежність до тестової групи (`ad` або `psa`)
- Чи здійснив користувач покупку (конверсія)
- Кількість переглянутих рекламних оголошень
- День і година найбільшого перегляду реклами

## **Мета аналізу**
- Обчислити та порівняти рівень конверсії для обох груп
- Виконати статистичний тест (χ²-тест) для перевірки значущості різниці в конверсії
- Оцінити вплив перегляду реклами на конверсію
- Побудувати візуалізації для наочного представлення результатів
